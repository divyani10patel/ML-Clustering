```python
x= df
x
```

```python
plt.scatter(x=df["Age"],y=df["Purchased"])
```

```python
kmeans= KMeans(n_clusters=17)
kmeans.fit(x)
```

```python
plt.scatter(x=df["Age"],y=df["Purchased"],c=y,cmap="rainbow")
```

## Elbow Method

```python
kmeans= KMeans(n_clusters=3)
kmeans.fit(x)
kmeans.predict(x)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
k = np.arange(1,10)
wcss_list= []

for i in k:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
```

```python
wcss_list
```

```python
plt.scatter(k,wcss_list)
```

```python
plt.plot(k,wcss_list)
```
